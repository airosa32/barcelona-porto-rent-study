# **Filtro de Dados - & igual a AND, | igual a OR**

```python
# Filtro: SELECT * FROM aluguel_casa_retratado WHERE (total <= 4000 AND animal = "accept") AND (rooms = 2 OR rooms = 3)
filtro_casa_1 = aluguel_casa_retratado[(aluguel_casa_retratado["total"]<=4000) & (aluguel_casa_retratado["animal"]=="accept") & ((aluguel_casa_retratado["rooms"]==2)|(aluguel_casa_retratado["rooms"]==3))]


filtro_casa_1.head()
```

# **Filtro de Dados - Query**

```python
# Filtro: SELECT * FROM aluguel_casa_retratado WHERE (total <= 4000 AND animal = "accept") AND (rooms = 2 OR rooms = 3)
filtro_de_casa_2 = aluguel_casa_retratado.query("(total <= 4000 and animal == 'accept') and (rooms == 2 or rooms == 3)")


filtro_de_casa_2.head()
```

# **Agrupamento - Groupby, Agg e Reset_index**

```python
# Filtro: SELECT city, AVG(total) FROM aluguel_casa_retratado WHERE (total <= 4000 AND animal = "accept") AND (rooms = 2 OR rooms = 3) GROUP BY city

# Agrupa "cidades" pelo "total" e tira a suas media
agrupamento_cidade = filtro_de_casa_2.groupby(["city"])["total"].agg("mean").reset_index()


agrupamento_cidade.head()
```

```python
# Filtro: SELECT bathroom, rooms, AVG(total) FROM aluguel_casa_retratado WHERE (total <= 4000 AND animal = "accept") AND (rooms = 2 OR rooms = 3) GROUP BY bathroom, rooms

# Agrupa "bathroom" e "rooms" pelo total e tira sua media
agrupamento_quartos = filtro_de_casa_2.groupby(["bathroom", "rooms"])["total"].agg("mean").reset_index()


agrupamento_quartos.head()
```

# **Visualizando os Dados - Graficos**

```python
import seaborn as sns

ax = sns.barplot(data=agrupamento_quartos, x="rooms", y="total", color="blue")


ax.set(xlabel="Qtd de quartos", ylabel="Média do preço", title="Casas");
```

```python
ax = sns.barplot(data=agrupamento_quartos, x="bathroom", y="total", hue="bathroom")


ax.set(xlabel="Qtd de banheiro", ylabel="Média do preço", title="Casas");
```

```python
# Refiltrando para area ser menor ou igual a 100
agrupamento_area = filtro_de_casa_2[['area']].query("area <= 100")

# Grafico de disperção dos dados = Forma/distribuição normal + desvio padrão
sns.histplot(data=agrupamento_area, x="area")
```

```python
sns.histplot(data=filtro_de_casa_2, x="total", hue="city");
```

# **Escolhendo o Apartamento**

```python
aluguel_casa_porto = filtro_de_casa_2.query("city == 'Porto'")


# Quantidades de apartamentos para alugar por andar (Andar, Quantidade)
aluguel_casa_porto.groupby("floor").index.nunique().sort_values(ascending=False)
```

```python
agrupamento_andar_preco = aluguel_casa_porto.groupby("floor")["total"].agg("mean").reset_index()


agrupamento_andar_preco.head()
```

```python
sns.barplot(data=agrupamento_andar_preco, x="floor", y="total");
```
